```python
import torch 
import numpy as np 
import matplotlib.pyplot as plt
```

```python
def sigmoid_k(x,alpha=1):
    return 1/(1+torch.exp(-alpha * x))
```

```python
n=torch.arange(-5,5,step=0.5)
#y=torch.sigmoid(n)
y = sigmoid_k(n,alpha=2)
plt.plot(n,y.numpy())
plt.show()
```

```python
n=torch.arange(-5,5,step=0.5)
y=torch.sigmoid(n)
#y = sigmoid_k(n,alpha=1)
plt.plot(n,y.numpy())
plt.show()
```
